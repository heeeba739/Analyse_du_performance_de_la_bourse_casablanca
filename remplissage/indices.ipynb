```python
import pandas as pd
import numpy as np

# CHARGEMENT
df = pd.read_csv("D:\\rouge_file\\Analyse_du_performance_de_la_bourse_casablanca\\csv_clean\\Indices_CLEAN.csv")
df = df[df['Type'] != 'Type'].drop_duplicates().copy()

print(f'Shape apres nettoyage basique : {df.shape}')
print(f'Nulls avant traitement        : {df.isnull().sum().sum()}')
print()

# VARIATION_VEILLE — 1 seul null (0.00%)
# Remplissage par 0.0
# Logique : c est l indice "MASI INDUSTRIE AGRICOLE" a sa
# premiere seance d existence — pas de veille disponible.
# On remplace par 0.0 car aucune variation connue = variation nulle.
# Ce n est pas une donnee manquante, c est un cas limite metier.
avant = df['Variation_Veille'].isnull().sum()
df['Variation_Veille'] = df['Variation_Veille'].fillna(0.0)
apres = df['Variation_Veille'].isnull().sum()
print(f'[ZERO] {"Variation_Veille":25s} : {avant} nulls -> {apres} nulls restants')

print()

# RAPPORT FINAL
print('RAPPORT FINAL')
print(f'Shape finale         : {df.shape}')
print(f'Nulls restants total : {df.isnull().sum().sum()}')
if df.isnull().sum().sum() == 0:
    print('Aucun null restant — tableau 100% propre !')

df.to_csv('Indices_FINAL.csv', index=False, encoding='utf-8')
print(f'Exporte : Indices_FINAL.csv — {len(df):,} lignes')
```
