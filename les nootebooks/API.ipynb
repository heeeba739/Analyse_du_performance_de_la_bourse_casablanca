```python
"""
Collecte des données en temps réel
Bourse de Casablanca via Yahoo Finance
Délai : ~15-20 min
"""

import yfinance as yf
import pandas as pd
from datetime import datetime
import os
import time


# CONFIGURATION


DOSSIER_OUTPUT = "DATA_TEMPS_REEL"
os.makedirs(DOSSIER_OUTPUT, exist_ok=True)

# LISTE DES SYMBOLES


SYMBOLES = {
    "ATW.CS"  : "Attijariwafa Bank",
    "BCP.CS"  : "Banque Centrale Populaire",
    "BOA.CS"  : "Bank Of Africa",
    "CIH.CS"  : "CIH Bank",
    "CDM.CS"  : "Crédit du Maroc",
    "BMCI.CS" : "BMCI",
    "WAA.CS"  : "Wafa Assurance",
    "ATL.CS"  : "Atlanta",
    "IAM.CS"  : "Maroc Telecom",
    "GAZ.CS"  : "Afriquia Gaz",
    "LES.CS"  : "Lesieur Cristal",
    "TQM.CS"  : "Totalenergies Marketing Maroc",
    "ADH.CS"  : "Addoha",
    "RDS.CS"  : "Résidences Dar Saada",
    "MNG.CS"  : "Managem",
    "CMA.CS"  : "Ciments du Maroc",
    "HOL.CS"  : "Holcim Maroc",
    "LBV.CS"  : "Label Vie",
    "ALM.CS"  : "Aluminium du Maroc",
    "SNP.CS"  : "Sonasid",
    "COL.CS"  : "Cosumar",
    "OUL.CS"  : "Oulmès",
    "CSR.CS"  : "Centrale Danone",
}


# SNAPSHOT TEMPS RÉEL


def snapshot_temps_reel():

    print("=" * 60)
    print("   SNAPSHOT TEMPS RÉEL - BOURSE DE CASABLANCA")
    print(f"   {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
    print("=" * 60)

    resultats = []

    for symbole, nom in SYMBOLES.items():
        try:
            ticker = yf.Ticker(symbole)
            info   = ticker.fast_info

            variation = round(
                ((info.last_price - info.previous_close) / info.previous_close) * 100, 2
            ) if info.previous_close else None

            resultats.append({
                "Date_Heure"       : datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
                "Symbole"          : symbole,
                "Societe"          : nom,
                "Cours"            : info.last_price,
                "Ouverture"        : info.open,
                "Plus_Haut"        : info.day_high,
                "Plus_Bas"         : info.day_low,
                "Cloture_Veille"   : info.previous_close,
                "Variation_%"      : variation,
                "Volume"           : info.last_volume,
            })

            signe = "GREEN" if variation and variation > 0 else "RED"
            print(f"  {signe} {nom:30s} | {info.last_price} MAD | {variation}%")
            time.sleep(0.3)

        except Exception as e:
            print(f"    {nom:30s} | Non disponible")

    # Sauvegarder
    if resultats:
        df = pd.DataFrame(resultats)

        # 1. Fichier horodaté (historique des snapshots)
        nom_horodate = f"snapshot_{datetime.now().strftime('%Y%m%d_%H%M%S')}.csv"
        df.to_csv(os.path.join(DOSSIER_OUTPUT, nom_horodate), index=False, encoding="utf-8")

        # 2. Fichier "dernier snapshot" toujours écrasé
        df.to_csv(os.path.join(DOSSIER_OUTPUT, "DERNIER_SNAPSHOT.csv"), index=False, encoding="utf-8")

        print("\n" + "=" * 60)
        print(f"   {len(resultats)} sociétés collectées")
        print(f"   Sauvegardé : {nom_horodate}")
        print("=" * 60)

    return df if resultats else None


# MODE AUTOMATIQUE (toutes les X minutes)

def lancer_en_continu(intervalle_minutes=20):
    """
    Lance un snapshot toutes les X minutes automatiquement
    Arrêt : Ctrl+C
    """
    print(f"\n   Mode continu activé — snapshot toutes les {intervalle_minutes} min")
    print("   Arrêt : Ctrl+C\n")

    while True:
        snapshot_temps_reel()
        print(f"\n   Prochain snapshot dans {intervalle_minutes} minutes...\n")
        time.sleep(intervalle_minutes * 60)


# MAIN

if __name__ == "__main__":

    print("\n  Que voulez-vous faire ?")
    print("  1 - Un seul snapshot maintenant")
    print("  2 - Mode continu (toutes les 20 min)\n")

    choix = input("  Votre choix (1/2) : ").strip()

    if choix == "1":
        snapshot_temps_reel()
    elif choix == "2":
        lancer_en_continu(intervalle_minutes=20)
    else:
        print("  Choix invalide.")
```
